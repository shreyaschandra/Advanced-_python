```python
# Introduction to MATPLOTLIB

-- Numpy  ---> numerical computation
-- Pandas  ---> data manuplation process
-- Matplotlib --> visually presentation of data
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
x=[1,2,3]
y=[3,5,7]
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
plt.plot(x,y)
plt.title('Test graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
z=[10,3,11]

plt.plot(x,y)
plt.plot(x,z)
plt.title('Test graph 1')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
#legend  is used represent the curve in the graph, it accepts data in list form.
plt.plot(x,y)
plt.plot(x,z)
plt.title('Test graph 1')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend(['X-Y graph','X-Z graph'])
plt.show()
```

```python
#working on data

sd=pd.read_csv('sample_data (2).csv')
sd
```

```python
type(sd)
```

```python
#plotting col a and col b

plt.plot(sd.column_a,sd.column_b)
plt.title('sample data')
plt.xlabel('column a')
plt.ylabel('column b')
plt.show()
```

```python
#plotting col a and col c

plt.plot(sd.column_a,sd.column_c)
plt.title('sample data')
plt.xlabel('column a')
plt.ylabel('column c')
plt.show()
```

```python
#column a, b, c

plt.plot(sd.column_a,sd.column_b)
plt.plot(sd.column_a,sd.column_c)
plt.title('sample data')
plt.legend(['A-B graph','A-C graph'])
plt.xlabel('column a')
plt.ylabel('column b and c')
plt.show()
```

```python
#Dot graph.

plt.plot(sd.column_a,sd.column_b,'o')
plt.title('sample data')
plt.xlabel('column a')
plt.ylabel('column b')
plt.show()
```
